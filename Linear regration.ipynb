```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/momen/Downloads/canada_per_capita_income.csv")
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
df
```

```python
df.head(5)
```

```python
plt.xlabel('Year')
plt.ylabel('Income(US$)')
m = plt.scatter(df.year,df.income, color ='RED')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['income'])
```

```python
prediction = reg.predict(np.array(2016).reshape(-1, 1))

# Print the prediction
print(prediction)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
predicted_income = reg.predict(df[['year']])

# Plotting the actual income against the predicted income
plt.plot(df['income'], predicted_income)

# Adding labels and title
plt.xlabel('Actual Income (US$)')
plt.ylabel('Predicted Income (US$)')
plt.title('Actual vs Predicted Income')

# Displaying the plot
plt.show()
```
